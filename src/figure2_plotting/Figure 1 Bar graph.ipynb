```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
labels = ['NCBI Genome','EMBL Metagenomic']
df = pd.read_csv('Tn7_subsystem_summary.csv',sep='\t')
NCBI_sum = df['NCBI'].sum()
META_sum = df['META'].sum()
NCBI_factor = NCBI_sum/100
META_factor = META_sum/100
Type_IF = np.array([df['NCBI'][0]/NCBI_factor,df['META'][0]/META_factor])
Type_IB =  np.array([df['NCBI'][1]/NCBI_factor,df['META'][1]/META_factor])
Type_IC = np.array([df['NCBI'][2]/NCBI_factor,df['META'][2]/META_factor])
Splits_protein= np.array([df['NCBI'][3]/NCBI_factor,df['META'][3]/META_factor])
Fusion_protein= np.array([df['NCBI'][4]/NCBI_factor,df['META'][4]/META_factor])
Type_IV= np.array([df['NCBI'][5]/NCBI_factor,df['META'][5]/META_factor])
Type_II= np.array([df['NCBI'][6]/NCBI_factor,df['META'][6]/META_factor])
#change the input
width = 0.35    
fig, ax = plt.subplots(figsize=(7,6))

ax.bar(labels, Type_IF,width,  label='Type IF',color='#56B4E9')
ax.bar(labels, Type_IB,width, bottom=Type_IF, label='Type IB',color='#0072B2')
ax.bar(labels, Type_IC,width, bottom=Type_IF+Type_IB, label='Type IC',color='#F0E442')
ax.bar(labels, Splits_protein,width,bottom=Type_IF+Type_IB+Type_IC, label='Splits protein',color='#009E73')
ax.bar(labels, Fusion_protein,width, bottom=Type_IF+Type_IB+Type_IC+Splits_protein, label='Fusion protein',color='#D55E00')
ax.bar(labels, Type_IV,width,bottom=Type_IF+Type_IB+Type_IC+Splits_protein+Fusion_protein, label='Type IV',color='#F0E442')
ax.bar(labels, Type_II,width,bottom=Type_IF+Type_IB+Type_IC+Splits_protein+Fusion_protein+Type_IV, label='Type II',color='gray')
ax.set_ylabel('Percentage',fontsize=14)
#ax.set_title('Scores by group and gender')
ax.legend(fontsize=14,loc='best')# check how to put the legend on the right

plt.xticks(size = 14)
plt.yticks(size = 14)
#plt.save('figure.png')
plt.show()
```
