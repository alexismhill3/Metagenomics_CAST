```python
import pandas as pd

df = pd.read_csv('./genefinder.csv',sep=',',names = ['contigID','region ','gene','position','id','direction','target_number','e','description'
                                             ,'aa_sequence','bitscore','rawscore','alignment_length','percent_identical'
                                             ,'number_identical','number_of_mismatches','positive_scoring','gap_openings','gaps'
                                             ,'percent_positive_matches','alignment_query_coverage','input_filename'])
```

```python
from Bio.SeqFeature import SeqFeature, FeatureLocation
#for genebank
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio import SeqIO
import gzip
from Bio import SeqIO
from Bio.Alphabet import IUPAC
def ths_pos(num1,num2):
    if num1 <= num2:
        return num1-1 , num2 , 1
    else:
        return num2-1, num1 ,-1
    
    


def get_feature(df):
    features_list =[]
    for x in range(len(df)):
        position = df['position'][x]
        pos_not_sorted_str = position.split('..')
        [num1,num2]= pos_not_sorted_str
       
        num1,num2 ,direction = ths_pos(int(num1),int(num2))
        pos_front = num1
        pos_back = num2

        gene = df['gene'][x]
        label = {}
        if gene == 'CRISPR array':
            label['label'] = 'CRISPR'
            feature = SeqFeature(FeatureLocation(start=pos_front, end=pos_back , strand=direction), type='mRNA',qualifiers=label)
        else:
            label['label'] = gene
            feature = SeqFeature(FeatureLocation(start=pos_front, end=pos_back,strand=direction), type='CDS',qualifiers=label)
        features_list.append(feature)
    return features_list

def get_seq_with_feature(id_name):
    df_test = df.loc[df['contigID'] == id_name]
    df_test=df_test.reset_index()
    features_list = get_feature(df_test)
    file = df_test['input_filename'][0]
    #print(df_test.head())
    src = file
    #print(src)
    records= list(SeqIO.parse(src, "fasta"))
    
    for x in records:
        if x.id == id_name:
            sequence = Seq(str(x.seq), IUPAC.unambiguous_dna)

    rec1 = SeqRecord(sequence,id=id_name,name=id_name,description=id_name)
    rec1.features = features_list
    return rec1
```

```python
#Write a genebank file
from Bio.SeqFeature import SeqFeature, FeatureLocation
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio import SeqIO
import gzip
from Bio import SeqIO
from Bio.Alphabet import IUPAC
from Bio.Alphabet import IUPAC
for x in ffs:

    id_name = x.split('-')[0]
    seqfront = int(x.split('-')[1])
    seqback = int(x.split('-')[2][:-4])
    print(id_name)
    the_seq = get_seq_with_feature(id_name)
    #print(id_name,the_seq)
    #rec1 = SeqRecord(the_seq,id=id_name,name='Example',description=id_name)
    name = './genebank/'+id_name+'-'+str(seqfront)+'-'+str(seqback)+'.gb'
    
    output_file = open(name, 'w')
    try:
        SeqIO.write(the_seq, output_file, 'genbank')
    except:
        the_seq.id= 'unknow'
        the_seq.name= 'unknow'
        the_seq.description= 'unknow'
        SeqIO.write(the_seq, output_file, 'genbank')
    output_file.close()
```
