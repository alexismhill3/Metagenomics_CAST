```python
import pandas as pd

df = pd.read_csv('/stor/work/Wilke/crisprtn7/NCBI_gene_finder.csv',sep=',',names = ['contigID','region ','gene','position','id','direction','target_number','e','description'
                                             ,'aa_sequence','bitscore','rawscore','alignment_length','percent_identical'
                                             ,'number_identical','number_of_mismatches','positive_scoring','gap_openings','gaps'
                                             ,'percent_positive_matches','alignment_query_coverage','input_filename'])
```

```python
import os
def Get_all_file(path):
    paths,files = [],[]
    # r=root, d=directories, f = files
    for r, d, f in os.walk(path):
        for file in f:    
            #print (file)
            p = os.path.join(r, file)
            paths.append(os.path.abspath(p))
            files.append(file)
    for f in files:
        #print(f)
        pass
    return paths, files



path = './typeI-B_png'
paths_flank, files_flank = Get_all_file(path)
```

```python
#for fasta
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio import SeqIO
import gzip
from Bio import SeqIO
from Bio.Alphabet import IUPAC
def get_seq(id_name,seqfront,seqback):

    df_test = df.loc[df['contigID'] == id_name]
    df_test=df_test.reset_index()
    file = df_test['input_filename'][0]

    relative_path = file.split("/scratch/07227/hilla3/projects/CRISPR-Transposons/data/genomic/NCBI")[-1]
    src = '/stor/work/Wilke/contig/database/metagenomic_contig_database/NCBI_redownload/genbank_bacteria_20210505'+relative_path
    #print(src)

    with gzip.open(src, "rt") as handle:
        records= list(SeqIO.parse(handle, "fasta"))
    #print('step1')
    for x in records:
        if x.id == id_name:
            sequence = Seq(str(x.seq), IUPAC.unambiguous_dna)
            #print(sequence)
    #sequence[seqfront:seqback]
    #print('step')
    return sequence[seqfront:seqback]

    
```

```python
#Write a fasta file
from Bio.Alphabet import IUPAC

def main(num):
    try:
        x = files_flank[num]
        id_name = x.split('-')[0]
        seqfront = int(x.split('-')[1])
        seqback = int(x.split('-')[2][:-4])
        print(id_name,seqfront,seqback)
        
        the_seq = get_seq(id_name,seqfront,seqback)
        
        rec1 = SeqRecord(the_seq,id=id_name,name='Example',description=id_name)
        name = './fasta/'+id_name+'-'+str(seqfront)+'-'+str(seqback)+'.fasta'
        SeqIO.write([rec1], name, "fasta")
    except:
        print(x = files_flank[num])
```

```python
def main1(num):
    try:
        main(num)
    except:
        pass
```

```python
from multiprocessing import Pool

if __name__ =='__main__':
    p = Pool(80)
    num = range(len(files_flank))
    #um=[0,1]
    p.map(main1,num)
    p.close()
    p.join()
```

```python
import ntpath
def path_leaf(path):
    head, tail = ntpath.split(path)
    return tail or ntpath.basename(head)
```
