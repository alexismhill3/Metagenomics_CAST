```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df  = pd.read_csv('typeI-F_att_site_summary.csv',sep=',')

labels = ['NCBI','Meta']
sum_NCBI = df['NCBI'].sum()
sum_Meta =df['Meta'].sum()
yciA = np.array([df['NCBI'][0]/sum_NCBI*100,df['Meta'][0]/sum_Meta*100])
guaC =  np.array([df['NCBI'][1]/sum_NCBI*100,df['Meta'][1]/sum_Meta*100])
rsmJ = np.array([df['NCBI'][2]/sum_NCBI*100,df['Meta'][2]/sum_Meta*100])
ffs= np.array([df['NCBI'][3]/sum_NCBI*100 ,df['Meta'][3]/sum_Meta*100 ])
unknown= np.array([df['NCBI'][4]/sum_NCBI*100 ,df['Meta'][4]/sum_Meta*100 ])


width = 0.35       # the width of the bars: can also be len(x) sequence
#plt.figure(figsize=(20,7))
fig, ax = plt.subplots(figsize=(15,15))

ax.bar(labels, yciA,width,  label='yciA',color='#DCC6A4')
ax.bar(labels, guaC,width, bottom=yciA, label='guaC',color='#86AABA')
ax.bar(labels, rsmJ,width, bottom=yciA+guaC, label='rsmJ',color='#B69FBE')
ax.bar(labels, ffs,width,bottom=yciA+guaC+rsmJ, label='ffs',color='#B2CEDB')
ax.bar(labels, unknown,width, bottom=yciA+guaC+rsmJ+ffs, label='unknown',color='#C9C9C9')

ax.set_ylabel('Percentage',fontsize=30)
#ax.set_title('Scores by group and gender')
#ax.legend(fontsize=20)
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[::-1], labels[::-1], fontsize=30,loc='0')
plt.xticks(size = 20)
plt.yticks(size = 20)
plt.savefig('typeI-F att site plot.eps', format='eps')
plt.show()
```
